```python
#Reading the csv file
df = pd.read_csv(r"C:\Users\Sudipta\Downloads\zomato.csv.zip")

#First 5 row of the dataset
df.head()
```

```python
#Let's see how many rows and columns it has
df.shape
```

```python
#So our dataset has 51717 rows and 17 columns
```

```python
#Let's extract the name of the columns
df.columns
```

```python
#So in our dataset it has some columns which has no use and also a column named "menu_item" which has no data
#So we can delete those columns
df = df.drop(['url', 'address', 'phone', 'menu_item', 'dish_liked', 'reviews_list'], axis = 1)

#Let's see first 5 rows of the modified dataset
df.head()
```

```python
#Let's cheak some information about our dataset
df.info()
```

```python
#There are some duplicate values
#We have to drop them
df.drop_duplicates(inplace = True)
df.shape
```

```python
#Now we start cleaning our dataset
#Let's clean the rate column
df['rate'].unique()
```

```python
#We want to convert anything other than numeric values to null value and remove '/' from 'rate ' column
def handlerate(value):
    if(value=='NEW' or value=='-'):
        return np.nan
    else:
        value = str(value).split('/')
        value = value[0]
        return float(value)
    
df['rate'] = df['rate'].apply(handlerate)
df['rate'].head()
```

```python
#Here we want to remove null values from 'rate' column
df['rate'].fillna(df['rate'].mean(), inplace = True)
#Let's cheak how many null values are still there
df['rate'].isnull().sum()
```

```python
#Here we removing all the null values of the dataset
df.dropna(inplace = True)
df.head()
```

```python
#We notices from the dataset that location and listed_in(city) has similar kind of data
#Let's cheak which one has less number of distinct data
df['listed_in(city)'].unique()
```

```python
df['location'].unique()
```

```python
#So "listed_in(city)" and "location" almost has same data
#So we can drop any one of them
#Drop 'location'
df = df.drop(['location'], axis = 1)

#Some column in our dataset has a long string as name of that column.It will be easy to work with those columns if we replace name of those column by some short name
#Rename those columns

df.rename(columns = {'approx_cost(for two people)':'Cost2plates', 'listed_in(type)':'Type','listed_in(city)':'City'}, inplace = True)
df.head()
 
```

```python
#Cleaning of the Cost2plates column

df['Cost2plates'].unique()

 
```

```python
#Some data are with comma.We want to delete those commas
def handlecomma(value):
    value = str(value)
    if ',' in value:
        value = value.replace(',', '')
        return float(value)
    else:
        return float(value)
    
df['Cost2plates'] = df['Cost2plates'].apply(handlecomma)
df['Cost2plates'].unique()
```

```python
#cleaning the rest_type column
rest_types = df['rest_type'].value_counts(ascending  = False)
rest_types
```

```python
rest_types_lessthan500 = rest_types[rest_types<500]
rest_types_lessthan500
```

```python
#Storing rest_type less than 500 frequency as others
def handle_rest_type(value):
    if(value in rest_types_lessthan500):
        return 'others'
    else:
        return value
        
df['rest_type'] = df['rest_type'].apply(handle_rest_type)
df['rest_type'].value_counts()
        
```

```python
#Cleaning cuisines column
cuisines = df['cuisines'].value_counts(ascending  = False)


cuisines_lessthan100 = cuisines[cuisines<100]



def handle_cuisines(value):
    if(value in cuisines_lessthan100):
        return 'others'
    else:
        return value
        
df['cuisines'] = df['cuisines'].apply(handle_cuisines)
df['cuisines'].value_counts()
```

```python
#We want to determine in a city how many restaurents has online ordering system
df1 = df.groupby(['City','online_order'])['name'].count()
df1.to_csv('City_online.csv')
df1 = pd.read_csv('City_online.csv')
df1 = pd.pivot_table(df1, values=None, index=['City'], columns=['online_order'], fill_value=0, aggfunc=np.sum)
df1
```

```python
# Maximum restaurents has online ordering system


#Let's plot it
df1.plot(kind = 'bar', figsize = (15,8))
```

```python
#Now we want to determine in a city how many restaurents has book table feature
df2 = df.groupby(['City','book_table'])['name'].count()
df2.to_csv('City_booktable.csv')
df2 = pd.read_csv('City_booktable.csv')
df2 = pd.pivot_table(df2, values=None, index=['City'], columns=['book_table'], fill_value=0, aggfunc=np.sum)
df2
```

```python
#Notice that maximum restaurents don't have Booking Table feature

#Here is our plot
df2.plot(kind = 'bar', figsize = (15,8))
```

```python
#Now we want to understand the relation between rate and Type
#Here is our visualization
plt.figure(figsize = (14, 8))
sns.boxplot(x = 'Type', y = 'rate', data = df, palette = 'inferno')
```

```python
#So Buffet , Pubs and bars , Drinks and nightlife type restaurents has maximum rate while restaurents of types Delivery , Dine-out has competitively low rate.



#Now we want to grouping type of restaurents with respected to various cities

df3 = df.groupby(['City','Type'])['name'].count()
df3.to_csv('City_Type.csv')
df3 = pd.read_csv('City_Type.csv')
df3 = pd.pivot_table(df3, values=None, index=['City'], columns=['Type'], fill_value=0, aggfunc=np.sum)
df3
```

```python
# So some city don't have any pubs and bars

#Let's plot it
df3.plot(kind = 'bar', figsize = (36,8))
```

```python
#Here we want to determine number of votes corresponding to a city
df4 = df[['City', 'votes']]
df4.drop_duplicates()
df5 = df4.groupby(['City'])['votes'].sum()
df5 = df5.to_frame()
df5 = df5.sort_values('votes', ascending=False)
df5.head()
```

```python
#So Koramangala 7th Block has maximum number of votes

#Let's make a visualization of it
plt.figure(figsize = (15,8))
sns.barplot(df5.index , df5['votes'])
plt.xticks(rotation = 90)
```

```python
#Now want to make a visualization of top cuisines according to number of votes
df6 = df[['cuisines', 'votes']]
df6.drop_duplicates()
df7 = df6.groupby(['cuisines'])['votes'].sum()
df7 = df7.to_frame()
df7 = df7.sort_values('votes', ascending=False)
df7.head()
```

```python
#we are not interested in others in cuisines
#So we have to drop others
df7 = df7.iloc[1:, :]
df7.head()
```

```python
#So North Indian cuisines has maximum number of votes

#Here is our plot of various cuisines vs votes
plt.figure(figsize = (15,8))
sns.barplot(df7.index , df7['votes'])
plt.xticks(rotation = 90)
```
